```python
df = sns.load_dataset('iris'); df.head()
```

```python
fig = plt.figure(figsize=(12, 7), facecolor='olive')
fig.subplots_adjust(hspace=0.4, wspace=0.4)
colors = ['red', 'green', 'olive', 'blue', 'pink', 'orange']
for i, c in zip(range(1, 7), colors):
    ax = fig.add_subplot(2, 3, i)
    ax.plot(range(10), stats.norm(0, 1).rvs(10), label='curve' + str(i), color=c, alpha=0.7, marker='*', linewidth=0.5)
    ax.axhline(0, color='grey', linestyle='--', label='horizontal')
    ax.axvline(3, color='#252525', linestyle='-.', label='vertical')
    ax.set_title('The Title')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_ylim(-2,1)
    ax.legend(loc='lower right')
```

```python
sns.lmplot(x="sepal_width", y="sepal_length", hue="species", data=df, 
                     height=7, aspect=1.5, palette='tab10', 
                     scatter_kws=dict(s=50, linewidths=0.7, edgecolors='black'));
```

```python
grid = sns.PairGrid(df, hue="species", height=3, aspect=1.5, palette='husl', diag_sharey=False,
                   vars=['sepal_length', 'sepal_width', 'petal_length'])
grid.map_diag(sns.kdeplot, shade=True)
grid.map_lower(sns.regplot)
grid.map_upper(sns.kdeplot)
grid.add_legend(title="Species");
```
