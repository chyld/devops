### Manual Subplots

```python
fig = plt.figure(figsize=(7, 2))
ax1 = fig.add_subplot(1, 2, 1)  # create a 1x2 - place at pos 1
ax1.scatter([1,2,3], [4,5,6])
ax2 = fig.add_subplot(1, 2, 2)  # create a 1x2 - place at pos 2
ax2.scatter([1,2,3], [4,5,6]);
```

### Auto Subplots

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 7), facecolor='olive')
colors = ['black', '#ff0000', 'olive', 'blue', '#ff33cc', 'orange']
markers = ['+', '*', '^','d', 'x', '.']
styles = ['solid', 'dotted', 'dashed', 'dashdot', 'solid', 'dotted']
for (i, ax), c, m, s in zip(enumerate(axs.flatten()), colors, markers, styles):
    ax.text(0.7, 1.3, "$f(x)=3^2 + x_1$", color=c, fontsize=15, bbox={'facecolor': 'red', 'alpha': 0.2, 'pad': 3})
    ax.plot(np.linspace(1, 50, 20), np.random.rand(20), color=c, label=c, alpha=0.7, marker=m, linewidth=0.7, linestyle=s)
    ax.axhline(0, color='grey', label='horizontal', alpha=0.3)
    ax.axvline(15, color='#252525', label='vertical', alpha=0.3)
    ax.set_title('The Title')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_ylim(-2,2)
    ax.legend(loc='lower right', frameon=False)
fig.tight_layout();
```

# Seaborn

```python
sns.lmplot(x="sepal_width", y="sepal_length", hue="species", data=df, 
                     height=7, aspect=1.5, palette='tab10', 
                     scatter_kws=dict(s=50, linewidths=0.7, edgecolors='black'));
```

```python
grid = sns.PairGrid(df, hue="species", height=3, aspect=1.5, palette='husl', diag_sharey=False,
                   vars=['sepal_length', 'sepal_width', 'petal_length'])
grid.map_diag(sns.kdeplot, shade=True)
grid.map_lower(sns.regplot)
grid.map_upper(sns.kdeplot)
grid.add_legend(title="Species");
```
